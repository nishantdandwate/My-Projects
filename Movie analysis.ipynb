```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
CompanyGrossSumSorted.plot(kind='bar').set(title='Total Earning By the Companies')
```

```python
# Looking at the top 15 director by gross revenue

CompanyGrossSum = df.groupby('director')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
CompanyGrossSumSorted.plot(kind='bar').set(title='Total Earning By the Directors')
```

```python
# this will replace all the object data type with a set of codes in numeric form 

df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
corr_matrix = df_numerized.corr(method='pearson')

sns.heatmap(corr_matrix, annot= True)
plt.title('Correlation Matrix for the Numeric feature')
plt.xlabel('Movies Feature')
plt.ylabel('Movies feature')
plt.show()
```
